### Q1. What are the most popular skills required for the role of a software developer/software engineer?


```python
skills = {} #started with empty dict.
for i in range(data.shape[0]): #will iterate 1000 times
    if 'software developer' in data['job_titles'][i] or 'software engineer' in data['job_titles'][i]:
        temp = data['skills'][i].split(",") #splitting data to list

        count = 0   #removing bracket coming in output, making the count accurate
        for ele in temp:
            if "[" in ele or "]" in ele:
                ele = ele.replace("[","")
                ele = ele.replace("]","")
            temp[count] = ele
            count += 1
        
        for ele in temp: 
            if ele not in skills: #if new skill is there, the frequency becomes one
                skills[ele] = 1
            else:
                skills[ele] += 1  #if the skill is already there, the frequency is incremented by 1 everytime until the loop
```

```python
skills
```

```python
n_skills = {"skills": skills.keys(), "Count": skills.values()} #creating new dict from skills to convert it to dataframe

new = pd.DataFrame.from_dict(n_skills)
```

```python
result = new.sort_values(by = "Count", ascending = False) #sorting the values in descending order
```

### Q4. Can you find some more interesting observations from this dataset?

```python
def bracketrmv(li):
    count = 0
    for e in li:
        if '[' in e or ']' in e:
            e = e.replace('[','')
            e = e.replace(']','')
        li[count] = e
        count += 1
    return li
```

```python
all_skills = result
```

```python
results1 = {}
for ele in all_skills["skills"]:
    results1[ele] = []
    for i in range(data.shape[0]):
        if ele in data['skills'][i]:
            for j in bracketrmv(data['job_titles'][i].split(",")):
                if j not in results1[ele]:
                    results1[ele].append(j)
                    
```

```python
results1
```

```python
for ele in results1:
    print("No Of Job Roles requiring {}: {}".format(ele,len(results1[ele])))
```

```python
val = 0
max_ele = ""
for ele in results1: #Linear Search
    if len(results1[ele]) > val:
        val = len(results1[ele])
        max_ele = ele
print("\n Most Popular Skill: {} with {} jobs requiring it".format(max_ele,val))
```

```python
#Finding the job with the most number of skills required

jobs = [] #Creating a list of all possible jobs
for i in range(data.shape[0]):
    for j in bracketrmv(data["job_titles"][i].split(",")):
        if j not in jobs:
            jobs.append(j)
```

```python
#Not the most efficient algorithm to find this, may have just used previous variables(?) 

srfj = {} #Skills required for each job
for job in jobs:
    srfj[job] = []
    for i in range(data.shape[0]):
        for j in bracketrmv(data["job_titles"][i].split(",")):
            if job in j:
                for k in bracketrmv(data["skills"][i].split(",")):
                    if k not in srfj[job]:
                        srfj[job].append(k)
```

```python
for job_title in srfj:
    print("\n Number of skills required for the role of {}: {}".format(job_title,len(srfj[job_title])))
```

```python
max = 0
job_title_max = ""
for job_title in srfj:
    if len(srfj[job_title]) > max:
        max = len(srfj[job_title])
        job_title_max = job_title
print("\n The role of {} requires the highest number of skills, with the total being {}".format(job_title_max,max))
```
